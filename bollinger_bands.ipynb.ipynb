```python
import pandas as pd
import numpy as np
import yfinance as yf
import datetime as dt
import matplotlib.pyplot as plt
```

```python
#download data from yfinnance.
df = yf.download(tickers='AAPL', start='2024-01-10', end=None)
print(df)
```

```python
#take mean of closing value then add 2 and subtract 2 from the mean to get bollinger bands.
df['mean'] = df['Close'].rolling(window=10).mean()
df['sd'] = df['Close'].rolling(window=10).std()

df['bb_hi'] = df['mean'] + (2*df['sd'])
df['bb_lw'] = df['mean'] - (2*df['sd'])
plt.plot(df['Close'], color='b')
plt.plot(df['bb_hi'], color='r')
plt.plot(df['bb_lw'], color='g')
plt.show()
```
